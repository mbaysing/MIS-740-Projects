# Teacher Salary Calculator
This program will ask the user to enter the starting salary for the first year and the number of years in teaching experience and for each year will output a salary. 

Input: Starting salary and number of years of experience teaching.

Output: Years of experience teaching and the salary for that amount of experience.

```python
#Begin the program.
import math #Import math library.

print('The program calculates the yearly salary for a teacher in ENSD.')
#Have user input their salary if not a valid number will prompt for them to put it in again.
print('What is the starting salary?')
startingSalary=float(input())
s1=startingSalary
while startingSalary <= 0: #I created a while loop so that it will continue to ask for a new variable until a valid number is put in place.
    print('The Salary should be a positive number. Please enter again:')
    startingSalary=int(input())

#User will input how many years of experience they have. If under one year will prompt them to re-enter it. 
print('How many years of experience in teaching?')
numOfYears=math.floor(float(input()))

while numOfYears < 1: #I created a while loop so that it will continue to ask for a new variable until a valid number is put in place.
    print('The years of experience should be at least 1. Please enter again:')
    numOfYears=float(input())
numOfYears=int(numOfYears)

for numOfYears in range(1, numOfYears+1): 
    if numOfYears == 1:
        s1=startingSalary
    elif numOfYears <=13:
        startingSalary *= (1.032)
        s1=startingSalary
    else:
        s1=s1
    print(f"For year {numOfYears:.0f} the salary is ${s1:,.2f}")
```
