## Tea Shop Ordering System
This program will 

Input: The user will input the item that they would like to order and a potential coupon code.

Output: The menu, selected items, and price of the transaction.

```python
#define the dictionary. The key is the menu item name, and the value is the price.
shopMenu={
    'Taro Bubble Tea': 8, 'Oolong Tea': 7, 'Tea Meets Coffee': 8, 'Lemongrass Tea': 8, 'Veggie Gua Bao': 10, 'Cheese Burger': 13, 'Tamago Sando': 10
}

#define the coupon string
validCoupons= ('welcome','old friend', 'until again', 'red blue and white', 'second chance', 'service')

#Print the opening statement and the menu
print('Welcome to Happy Tea Shop! Please order from the following items:')
for k,v in shopMenu.items():
    print(k.ljust(20)+ str(v).rjust(4))

#makes dictionary user input case insensitive. I had to use chatgpt to understand this concept better. 
#I used it to understand the k for k in shopMenu as I kept doing item.lower() in multiple places and could not get it to work.
lowershopMenu= {k.lower(): k for k in shopMenu} 
#prompt the user to enter a menu item
cart={}

while item !=" ":
    print('\nPlease enter the item you need. Hit enter (blank) to stop:\n')
    item=input()
    item_lower=item.lower()
    
    if item_lower in lowershopMenu:
        item=lowershopMenu[item_lower]
        
    if item in cart:
        print(f'The service "{item}" is already in the cart')
    elif item in shopMenu: #keeps duplicates from being added
        cart[item]=shopMenu[item]
        print('\nAdded to the cart')
    elif item == '':   #ends the loop if enter is hit
        break
    else:
       print(f"We do not offer \'{item.title()}\'")

if cart:
    total= sum(cart.values())      #calculate total cost
    
#make coupon case insensitive as well as asking for user to input their coupon
    coupon=input('\nPlease enter coupon code if you have one:\n').lower()
    if coupon != '' and coupon.lower() not in (c.lower() for c in validCoupons):
        print('\nCannot find the coupon code you entered:', coupon,'.\n')
        
    def printItemList(cart, leftWidth, rightWidth):      #print cart and show the total charge
            print('\nThank you for placing the order. The items you selected are as following:\n')
            for k, v in cart.items():
                print(k.ljust(leftWidth, '.') + f'${v:,.0f}'.rjust(rightWidth))
    printItemList(cart, 30, 4)

    print(f'\nThe total charge of the items you ordered is ${total:,.2f}\n')

#determine if user has a valid coupon to discount the total by 10%
    if coupon=='':
        print('\nYou did not have our coupon? Here is one for your next order. Please enter \"welcome\" next time when you visit us.\n')
    elif coupon in (c.lower() for c in validCoupons):
        discountedtotal=total*.9
        print('\nWe will provide a 10% discount for your order (coupon code:', coupon, ').\n') 
        print(f'The total price is {discountedtotal:,.2f}')
    else: 
        print('\nYou did not have our coupon? Here is one for your next order. Please enter \"welcome\" next time when you visit us.\n')

#If user does not input an item into the cart it will display this error message.
else:
    print('You did not select any item. Sorry we cannot prepare the order for you.')
```
