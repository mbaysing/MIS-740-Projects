```python
#Referenced the Flow Chart solution .pdf file from class for this code.
#print the purpose of this program
print('The purpose of the program is to calculate the service fee for a stock trading company.')

#Have user input the transaction amount from the last month
print('Please enter the transaction amount last month:')

transaction= float(input())
if transaction < 0:
    print('Cannot generate fee from a negative transaction amount. Please input a valid transaction amount.') 
#Determine if the transaction amount requires a fee % or a standard amount
elif transaction == 0:
    fee= 5
    print(f" The service fee for the customer is ${fee:,.2f}")  
else:
    #Have user input the year the account was opened
    print('Please enter the year the account was opened:')
    year= int(input())
    if year <= 0:   #imbedded if/else statement into else so that the year input prompt does not occur if the transaction is not in the range for it.
        print('Year cannot be a negative number or 0, please input a valid year.')  #Also keeps from generating a false service fee if the transaction is negative
    else:
        #Determine the fee % based on the year 
        if  0 < year <= 2010:
            fee= transaction*0.015
        else:
            fee= transaction*0.019
            if fee >= 1000:  #to get fee to display as $1,000 if calculated at $1,00 or more
                fee= 1000
            else:
                fee= fee #allows the original fee amount to print. 
        print(f" The service fee for the customer is ${fee:,.2f}")  #In line with prior if/else statement so that it will print for both year questions. 
```
