# BMI Calculator
The user enters the height in inches and weight in pounds, which will be converted to meters and kilograms to calculate the BIM.

Input: height and weight
Output: BMI
Author: Michaela Baysinger
Date: 09/11/2025

```python
# Formula for BMI is weight in kilograms/ (height in meter)^2
# Define the constants in ALL CAPS in other languages this is a constant
# Make sure to create constants instead of putting the specific number next time
#Get user input for weight in pounds
print('Please enter the weight in pounds: ')
weight= float(input())

# Get user input for height
print('Please enter the height in inches: ')
height= float(input())

#Convert weight from pounds to kilograms
weight= weight/2.205

#Convert height from inches to meters
height= height/39.37

#Calculate BMI
BMI= weight/(height**2)

print(f"Your BMI is {BMI:.1f} kg/m^2")
```
