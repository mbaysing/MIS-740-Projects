## Assignment 5 Energy Efficiency Analysis
The Environmental Department collected household energy usage data from California (CA) and Nevada (NV). Each record represents one householdâ€™s monthly usage across three categories: electricity, natural gas, and water.

Input: The input comes from the two dataset files: usage_CA.csv and usage_NV.csv which has household energy usage in the three categories: electricity, gas, and water.

Output: The summary statistics for the total impact score, specifically: mean, standard deviation, and median. It will also display the number of households in each tier of environmental impact.

```python
import numpy as np
import pandas as pd
#calling csv for California usage and checking that it is correct
ca= pd.read_csv('usage_CA.csv')
#ca.head(5)

#printing the number of households in the California records
print(f'There are {ca.shape[0]:,} records for California households.')

#Creating Arrays for each category of energy usage for California and getting the total records in each category
caElectricityArray= np.array(ca['electricity'])
#print(caElectricityArray[:5])
#print(caElectricityArray.shape)

caGasArray= np.array(ca['gas'])
#print(caGasArray[:5])
#print(caGasArray.shape)

caWaterArray= np.array(ca['water'])
#print(caWaterArray[:5])
#print(caWaterArray.shape)

#Creating Arrays for each category of energy usage for Nevada and getting the total records in each category

nv= pd.read_csv('usage_NV.csv')
#nv.head(5)

#printing the number of households in the Nevada records
print(f'There are {nv.shape[0]:,} records for Nevada households.')

#Creating Arrays for each category of energy usage for Nevada

nvElectricityArray= np.array(nv['electricity'])
#print(nvElectricityArray[:5])
#print(nvElectricityArray.shape)

nvGasArray= np.array(nv['gas'])
#print(nvGasArray[:5])
#print(nvGasArray.shape)

nvWaterArray= np.array(nv['water'])
#print(caWaterArray[:5])
#print(nvWaterArray.shape)

#Combine both California and Nevada electricity arrays into one
ElectricityArray= np.concatenate([caElectricityArray, nvElectricityArray], axis=0)

#Calculate environmental impact of all households, because set to 1 just set it equal to ElectrityArray
electricity=ElectricityArray

#Combine both California and Nevada gas arrays into one
GasArray= np.concatenate([caGasArray, nvGasArray], axis=0)


#Combine both California and Nevada water arrays into one
WaterArray= np.concatenate([caWaterArray, nvWaterArray], axis=0)


totalusage=np.column_stack((ElectricityArray, GasArray, WaterArray))
print(f'{totalusage.shape[0]:,}')

#Calculate the environmental impact for all resources
#weight factors for electricity, natural gas, and water
weights=np.array([1,3,10]) 

#calculate environmental impact for all households by multiplying weight factor for each resource.
totalimpact=np.multiply(totalusage,weights)
totalimpact=np.sum(totalimpact, axis=1)
#Calculate mean, standard deviation, and median for the total impact score
mean_totalimpact=np.mean(totalimpact)
std_totalimpact=np.std(totalimpact)
median_totalimpact=np.median(totalimpact)

#Display summary statistics for the total impact score
print(f"{'The mean of the impact points is:':49}  {mean_totalimpact:,.2f}")
print(f"{'The standard deviation of the impact points is:':50} {std_totalimpact:,.2f}")
print(f"{'The median of the impact points is:':50} {median_totalimpact:,.2f}")

#How many households are in each tier of environmental impact
greenTier=np.sum(totalimpact < 1500)
print(f"{'Green Tier:':30} {greenTier:,.2f}")

averageTier=np.sum((totalimpact >= 1500) & (totalimpact < 3000))
print(f"{'Average Tier:':30} {averageTier:,.2f}")

highTier=np.sum(totalimpact >= 3500)
print(f"{'High Tier:':30} {highTier:,.2f}")

#Compare mean usage by state (CA vs NV) for electricity
caMeanElectricity= np.mean(caElectricityArray)
nvMeanElectricity= np.mean(nvElectricityArray)
print(f"\nThe mean usage of electricity in California is: {caMeanElectricity:,.2f}"
      f" compared to the mean usage of electricity in Nevada: {nvMeanElectricity:,.2f}\n")

#Compare mean usage by state (CA vs NV) for gas
caMeanGas= np.mean(caGasArray)
nvMeanGas= np.mean(nvGasArray)
print(f"\nThe mean usage of gas in California is: {caMeanGas:,.2f}"
      f" compared to the mean usage of gas in Nevada: {nvMeanGas:,.2f}\n")

#Compare mean usage by state (CA vs NV) for water
caMeanWater= np.mean(caWaterArray)
nvMeanWater= np.mean(nvWaterArray)
print(f"\nThe mean usage of water in California is: {caMeanWater:,.2f}"
      f" compared to the mean usage of water in Nevada: {nvMeanWater:,.2f}\n")

caArray=np.column_stack((caElectricityArray,caGasArray,caWaterArray))
caImpact=np.multiply(caArray,weights)
caImpact=np.sum(caImpact,axis=1)

caGreenTier=np.sum(caImpact < 1500)
print(f"{'California Green Tier:':30} {caGreenTier:,.2f}")
caAverageTier=np.sum((caImpact >= 1500) & (caImpact < 3000))
print(f"{'California Average Tier:':30} {caAverageTier:,.2f}")
caHighTier=np.sum(caImpact >= 3500)
print(f"{'California High Tier:':30} {caHighTier:,.2f}")

nvArray=np.column_stack((nvElectricityArray,nvGasArray,nvWaterArray))
nvImpact=np.multiply(nvArray,weights)
nvImpact=np.sum(nvImpact,axis=1)

nvGreenTier=np.sum(nvImpact < 1500)
print(f"{'Nevada Green Tier:':30} {nvGreenTier:,.2f}")
nvAverageTier=np.sum((nvImpact >= 1500) & (nvImpact < 3000))
print(f"{'Nevada Average Tier:':30} {nvAverageTier:,.2f}")
nvHighTier=np.sum(nvImpact >= 3500)
print(f"{'Nevada High Tier:':30} {nvHighTier:,.2f}")
```

The most important finding that I found was that California had higher mean usage of all three utilities than Nevada. Which prompted me to also checked to see the breakdown of the environmental impact tiers for California and Nevada. After doing that, it can be seen that California makes up the majority of the high tier and Nevada makes up the majority of the green tier. This correlates with the higher energy usage in California.
